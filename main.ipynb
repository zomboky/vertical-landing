```python
import time
```

Démarrage du moteur Pybullet 

```python

p.connect(p.GUI) 
```

```python
#Redéfinition du PATH pour le chargement des objets

p.setAdditionalSearchPath(pybullet_data.getDataPath())
```

```python
# Gravité
p.setGravity(0, 0, -9.81)

# Sol (plane)
plane_id = p.loadURDF("plane.urdf")
```

Création de la Fusée


```python
# Cyclindre visuel 
visual_shape = p.createVisualShape(p.GEOM_CYLINDER, radius=1.1, length=13.0, rgbaColor=[0.5, 0.5, 0.5, 1])
```

```python
# Collision du cylindre
collision_shape = p.createCollisionShape(p.GEOM_CYLINDER, radius=0.1, height=1.0)
```

```python
# Créer le corps (fusée)
mass = 1500 #Masse en kg
fusée_id = p.createMultiBody(
    baseMass=mass,
    baseCollisionShapeIndex=collision_shape,
    baseVisualShapeIndex=visual_shape,
    basePosition=[0,0,0.5]
)
```

Position de la Caméra de base au lancement du programme


```python
# Position de la caméra
p.resetDebugVisualizerCamera(
    cameraDistance=5,          # distance caméra
    cameraYaw=45,              # rotation horizontale
    cameraPitch=-30,           # rotation verticale
    cameraTargetPosition=[0,0,0]  # point regardé
)
```

Boucle d'actualisation


```python
# Boucle de simulation
for i in range(240):  # 1 seconde à 240 Hz
    p.stepSimulation()
    time.sleep(1./240.)
```
